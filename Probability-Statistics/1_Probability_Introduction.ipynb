## 1.1 What is Probability?

Probability quantifies how likely an event is to occur — a number between 0 (impossible) and 1 (certain).

**Example:**  
Flipping a fair coin:

- $P(\text{Heads}) = 0.5$
- $P(\text{Tails}) = 0.5$

---

## 1.2 Key Concepts

### 🎲 Experiment
An action that produces uncertain outcomes.  
_E.g., flipping a coin, rolling a die._

### 🧺 Sample Space ($\Omega$)
The set of all possible outcomes.  
_E.g., rolling a die: $\Omega = \{1, 2, 3, 4, 5, 6\}$_

### 🎯 Event
A subset of the sample space.  
_E.g., rolling an even number: $A = \{2, 4, 6\}$_

### 📏 Classical Probability (Equally likely outcomes)
$$
P(A) = \frac{\text{Number of favorable outcomes}}{\text{Total number of outcomes}}
$$

**Example:**  
$P(\text{even number}) = \frac{3}{6} = 0.5$

---

## 1.3 Rules of Probability

### ➕ Addition Rule
If two events A and B are mutually exclusive (can’t occur together):

$$
P(A \cup B) = P(A) + P(B)
$$

**Example:**  
$P(1 \text{ or } 2) = \frac{1}{6} + \frac{1}{6} = \frac{1}{3}$

---

### ✖️ Multiplication Rule
If two events A and B are independent:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

**Example:**  
Flip 2 coins.  
$P(\text{H and H}) = \frac{1}{2} \cdot \frac{1}{2} = \frac{1}{4}$

---

### 🔁 Complement Rule
Probability that an event does **not** happen:

$$
P(\text{not } A) = 1 - P(A)
$$

**Example:**  
$P(\text{not 6}) = 1 - \frac{1}{6} = \frac{5}{6}$

---

## 1.4 Conditional Probability

The probability of A given that B has occurred:

$$
P(A|B) = \frac{P(A \cap B)}{P(B)}
$$

**Example (Card deck):**

- $P(\text{ace and heart}) = \frac{1}{52}$
- $P(\text{heart}) = \frac{13}{52}$
- $P(\text{ace} | \text{heart}) = \frac{1/52}{13/52} = \frac{1}{13}$

---

## 1.5 Bayes’ Theorem

Bayes' Theorem helps reverse conditional probabilities:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

**Example (Medical Test):**

- $P(\text{Disease}) = 0.01$
- $P(\text{Positive} | \text{Disease}) = 0.95$
- $P(\text{Positive} | \text{No Disease}) = 0.05$

**Step 1: Compute $P(\text{Positive})$**

$$
P(\text{Positive}) = 0.95 \cdot 0.01 + 0.05 \cdot 0.99 = 0.0095 + 0.0495 = 0.059
$$

**Step 2: Compute $P(\text{Disease} | \text{Positive})$**

$$
P(\text{Disease} | \text{Positive}) = \frac{0.95 \cdot 0.01}{0.059} \approx 0.161
$$

> So even with a positive test, the actual chance of disease is ~16.1%.
